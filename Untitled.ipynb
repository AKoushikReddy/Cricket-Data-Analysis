# IPL Data Analysis

###### Have gathered IPL Statistics from all the seasons till 2022.

###### In an view that most of the T20 Matches are batting heavy , and toss dependent. Only batting statistics have been considered.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
sns.set()
```

```python
df = pd.read_csv('all_season_batting_card.csv')
```

```python
df.drop(['commentary','link','minutes','captain'],axis=1,inplace=True)
```

```python
convert_dict = {'match_id': str, 'current_innings':str}
df=df.astype(convert_dict)
```

```python
first_batting_records = df.loc[df['innings_id']==1]
second_batting_records = df.loc[df['innings_id']==2]
```

```python
#first_batting_runs = pd.pivot_table(first_batting_records,values=['runs','match_id'],index=['season'],columns=['current_innings'],aggfunc=np.sum, fill_value=0)
```

```python
#first_batting_innings = pd.pivot_table(first_batting_records,values=['match_id'],index=['season'],columns=['current_innings'],aggfunc=lambda x: len(x.unique()), fill_value=0)
```

```python
first_batting_innings=first_batting_records.groupby(['season','current_innings',])['match_id'].nunique()
second_batting_innings=second_batting_records.groupby(['season','current_innings'])['match_id'].nunique()
```

```python
first_batting_x=[i[0] for i in first_batting_innings.index]
second_batting_x=[i[0] for i in second_batting_innings.index]
```

```python
first_batting_runs=first_batting_records.groupby(['season','current_innings'])['runs'].sum()
second_batting_runs=second_batting_records.groupby(['season','current_innings'])['runs'].sum()
```

```python
first_batting_avg = first_batting_runs/first_batting_innings
second_batting_avg = second_batting_runs/second_batting_innings
```

```python
first_batting_season_avg=first_batting_avg.groupby(level=0).mean()
second_batting_season_avg=second_batting_avg.groupby(level=0).mean()
```

```python
summary = pd.read_csv('all_season_summary.csv')
df=summary.loc[:,['season','venue_name']].groupby(['season']).nunique()
```

#### First Innings & Second Innings Average

```python
fig, axs = plt.subplots(1)
fig.set_size_inches(12, 3, forward=True)
axs.plot(first_batting_season_avg.index,first_batting_season_avg.values-second_batting_season_avg.values)
axs.plot(first_batting_season_avg.index,df['venue_name'].values)
plt.legend(["Average Difference (1st batting - 2nd batting)", "Number of Venues"], loc ="upper left")
```

##### The above Line Graph, shows the difference in average for each season.
##### Its quite consistent until 2013.
##### From 2014 we can observe the change with respect to the change in number of venues
##### A low difference indicates that targets have been chased more often, and a high difference indicates that team's batting first have won.
##### when the number of venues have gone down we can see the hike in the difference and vice versa.This can be intrepreted as follows:
##### Its easier to bat second when there are more venues , as grounds are fresh and well mainted and difficult to chase when the number of venues decrease , more number of games played on the same ground , pitches degrade faster and makes batting second a disadvantage.

```python
summary = pd.read_csv('all_season_summary.csv')
```

```python
summary=summary.loc[:,['season','short_name','home_team','away_team','toss_won','decision','winner','venue_name']]
```

```python
def innings(x):
    if x.toss_won == 'CSK':
        if x.decision == 'BAT FIRST':
            return 1
        else:
            return 2
    else:
        if x.decision == 'BAT FIRST':
            return 2
        else:
            return 1
```

```python
def game_won(x):
    if x.winner == x.Team:
        return 'WON'
    else:
        return 'LOST'
```

```python
def toss_won(x):
    if x.toss_won == x.Team:
        return 1
    else:
        return 0
```

```python
def overall(x):
    if x.innings_id == 1:
        if x.toss == 1:
            return 'WON_TOSS BAT'
        else:
            return 'LOST_TOSS BAT'
    else:
        if x.toss == 1:
            return 'WON_TOSS BOWL'
        else:
            return 'LOST_TOSS BOWL'
```

```python
ar = summary['home_team'].unique()
for Name in ar:
    Team = summary[(summary['home_team'] == Name) | (summary['away_team'] == Name)]
    factor=Team['season'].unique()
    x_axis = np.arange(len(factor))
    Team=Team[Team['season'] >2000]
    Team['Team'] = Name
    Team['innings_id']=Team.loc[:,['decision','toss_won']].apply(innings,axis=1)
    Team['toss'] = Team.loc[:,['toss_won','Team']].apply(toss_won,axis=1)
    Team['win'] = Team.loc[:,['winner','Team']].apply(game_won,axis=1)
    Team['overall']=Team.loc[:,['innings_id','toss']].apply(overall,axis=1)
    df1=Team[Team['win'] == 'LOST']
    df1=df1.groupby(['season','toss']).size().reset_index().rename(columns = {0:'count'})
    g=sns.catplot(x='season', y='count', hue='toss', data=df1, kind='bar', height=6.5, aspect=10/7,ci=0).set(xlabel ="Season", ylabel = "Wins", title = Name)
    g._legend.set_title('TOSS RESULT')
    new_labels = ['Lost', 'Won']
    for t, l in zip(g._legend.texts, new_labels):
        t.set_text(l)
```

### Here is another bar plot which show the toss result on games which are lost by each team

#### we can observe that in most of the games lost , they could'nt win the toss.

#### It stregths the argument that, IPL sucess is dependent on ground conditions and toss.
#### Such factors are in every league , but IPL league has with very few teams ,with immense talent.So these external factors play out over the course of full season in IPL comparitively more than the other leagues.
